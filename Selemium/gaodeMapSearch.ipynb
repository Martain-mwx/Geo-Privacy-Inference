```python
from seleniumwire import webdriver
from msedge.selenium_tools import Edge, EdgeOptions
from selenium.webdriver.common.keys import Keys
import requests
import time

options = EdgeOptions()
options.use_chromium = True
# 无头模式
#options.add_argument('--headless')
options.add_experimental_option('useAutomationExtension',False)
options.add_experimental_option('excludeSwitches', ['enable-automation', 'enable-logging'])
options.add_argument("--disable-blink-features=AutomationControlled");
options.add_argument("--auto-open-devtools-for-tabs");
options.add_argument("--start-maximized");

# 设置好应用扩展 
extension_path = r"D:\文档资料\vscodeWork\gaodeMapHook\gaodeMapHook_v1.crx"
options.add_extension(extension_path)

options.binary_location = r"C:\Program Files (x86)\Microsoft\Edge\Application\msedge.exe" # 浏览器的位置
driver = Edge(options=options, executable_path=r"D:\tool\edgedriver_win64\msedgedriver.exe") # 相应的浏览器的驱动位置
url = "https://ditu.amap.com/"
driver.get(url)
```

```python
#以form表单形式向后台发送数据 application/x-www-form-urlencoded
def send_data(url,data):
    data = {'data':data}
    s = requests.session()
    s.keep_alive = False # 关闭多余连接
    s.post(url,data,timeout=300)
    
#延时刷新浏览器,并通知后台
def refresh(delay,url,localhost):
    time.sleep(delay)
    send_data(localhost,'refresh')
    driver.execute_script("window.location.href = " + '"'+ url +'"')

#延时delay秒，然后对目标地点进行搜索
def search(delay,location):   
    time.sleep(delay)
    #搜索框输入
    driver.find_element_by_id("searchipt").send_keys(location)
    u =  "console.log("+"'"+location+"'"+")"
    driver.execute_script(u)
    time.sleep(0.5)
    driver.find_element_by_xpath('//*[@id="searchbtn"]/i').click()
    

#切换到对应城市
def city(delay,city_num):
    time.sleep(delay)
    driver.find_element_by_xpath('//*[@id="citybox"]/div[5]').click()
    time.sleep(1)
    if city_num < 20:
        xpath = '//*[@id="citybox"]/div[7]/ul/li' + '[' + str(city_num) + ']'
        driver.find_element_by_xpath(xpath).click()
    elif city_num == 20:
        y = '//*[@id="citybox"]/div[7]/section[1]/div[2]/input'
        input = driver.find_element_by_xpath(y)
        input.send_keys('兰州')
        time.sleep(1)
        input.send_keys(Keys.ENTER)
    elif city_num == 21:
        y = '//*[@id="citybox"]/div[7]/section[1]/div[2]/input'
        input = driver.find_element_by_xpath(y)
        input.send_keys('呼和浩特')
        time.sleep(1)
        input.send_keys(Keys.ENTER)
    pass




def s(targetLocation,refreshUrl,localhost,city_num):
    send_data(localhost,targetLocation)
    #切换到对应城市
    city(1,city_num)
    #延时2s搜索
    search(3,targetLocation)
    refresh(7,refreshUrl,localhost)
```
